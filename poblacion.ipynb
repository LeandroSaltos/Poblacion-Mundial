```python
#Poblacio Mundial

#import numpy as np
import matplotlib.pyplot as plt

pob_1800=985


años=[x for x in range(1800,2021)]
pob=[pob_1800]

def modelacion_poblacion(tasas,años,poblacion):
    for i in range(len(años)-1):
        
        if años[i]<1945:
            
            x=poblacion[i]+poblacion[i]*tasas[0]
            poblacion.append(x)
        else:
            x=poblacion[i]+poblacion[i]*tasas[1]
            poblacion.append(x)
            
    return poblacion

m=modelacion_poblacion([0.0066,0.01], años, pob)

def TCAP(valores):
    
    v0=valores[0]
    vn=valores[-1]
    n=len(valores)
    
    t=((vn/v0)**(1/(n-1)))-1
    return t

tp=TCAP(m)
nueva_pob=[pob_1800]
for i in range(len(años)-1):
    x=nueva_pob[i]+nueva_pob[i]*tp
    nueva_pob.append(x)

plt.figure(figsize=(6,8))

plt.title('Poblacion Mundial 1800-2021')
plt.plot(años,m,'g',label='Datos Historicos')
plt.plot(años,nueva_pob,'r',label='Tasa Promedio Anual')


plt.legend()
plt.show()
```
